```python
#import library
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn.cluster import KMeans
```

```python
#input dataset

x = [1, 5, 1.5, 8, 1, 9]
y = [2, 8, 1.8, 8, 0.6, 11]

plt.scatter(x,y)
plt.show()
```

```python
#parse all dataset lines
with open('data/titanic.csv',encoding='utf8') as f:
    reader = csv.DictReader(f)
    data = [line for line in reader if line['age']]
#getting only coordinates
classes = [float(line['pclass'])*100 for line in data]
ages = [float(line['age']) for line in data]


plt.scatter(ages,classes)
plt.show()
```

```python
#wrappping dataset into np
points = []
for i in range(len(ages)):
    points.append([ages[i],classes[i]])
input_dataset_array = np.array(points)
```

```python
#kmeans computation here
kmeans = KMeans(n_clusters=4)
kmeans.fit(input_dataset_array)

#getting the results
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
colors = ["g.","r.","c.","y."]

for i in range(len(input_dataset_array)):
    #print("coordinate:",input_dataset_array[i], "label:", labels[i])
    plt.plot(input_dataset_array[i][0], input_dataset_array[i][1], colors[labels[i]], markersize = 10)


plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=150, linewidths = 5, zorder = 10)

plt.show()
```
